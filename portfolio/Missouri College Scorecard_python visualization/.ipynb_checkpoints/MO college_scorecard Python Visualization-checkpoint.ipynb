## 3. Upward Mobility of Missouri colleges

```python


ax = plt.scatter(\
            x=scorecard['first_bracket_net_price'], 
            y=scorecard['avg_income_from_low_income_bracket'], 
            s=scorecard['undergrad_enrollment']/10, # s is size
            alpha=0.4)  
plt.title("MO college earnings/cost")
plt.xlabel("first_bracket_net_price")
plt.ylabel("Earnings_from_low_income_bracket")
```

```python
ax = plt.scatter(\
            x=scorecard1['first_bracket_net_price'], 
            y=scorecard1['avg_earnings_from_low_income_bracket'], 
            s=scorecard1['undergrad_enrollment']/1000, # s is size
            alpha=0.4)  
plt.title("U.S.college earnings/cost")
plt.xlabel("first_bracket_net_price")
plt.ylabel("Earnings_from_low_income_bracket")
sns.regplot(x="first_bracket_net_price", y="avg_earnings_from_low_income_bracket", data=scorecard1, ax=g.axes[0, 0])
return_type = 'dict'
```

School cost and median earnings of Missouri students tend to cluster below $15000 and $40000 respectively, close to the national level. 

```python
sns.set(style="ticks")

g = sns.factorplot(x="compeletion_rate", y="median_earnings", hue="majority_black_flag", data=scorecard)
```

```python
sns.set(style="ticks")

g = sns.factorplot(x="compeletion_rate", y="median_earnings", hue="majority_pell_flag", data=scorecard)
```

Performance: Race and family income both play a role in the academic and economic performance of college students. Students in the Missouri colleges where majority are black students have lower completion rate and median earnings. Students in the Missouri colleges where majority are from low-income family have lower median earnings. We need to ask why and explore how to break the vicious circle. 

```python
g = sns.FacetGrid(scorecard, col="majority_pell_flag")  
g.map(sns.regplot, "avg_net_price", "median_earnings")  
```

```python
g = sns.FacetGrid(scorecard, col="majority_black_flag")  
g.map(sns.regplot, "avg_net_price", "median_earnings")  
```

Economic return of higher education: The net price of Missouri colleges is positively related with the earnings of graduates.Schools that accept more Pell recipients have lower post-graduation median earnings, and lower net price. Race does not play a significant role here.

## 4.  SQL check and interview(Seek out outliers and human voice, to be continued)

Based on the analysis of College Scorecard database released by the Obama administration, over 60% of Missouri universities charged the neediest students more than half of their family income, whose annual family income is less than $30,000. On the other hand, only one thirds of Missouri universities can help the poor students moved out of the lowest income tercile.


Washington University in St Louis and Saint Louis University, among all types of universities, stands out in the extremely low percentage of Pell Grants recipients and high cost toward low-income students. Washington University, with over 6 billion endowment and 7% Pell recipients, fell in the bottom 5 percent of all colleges in U.S. in its enrollment of Pell students. It’s also one of the 38 colleges that had more students from the top 1 percent than the bottom 60 percent, according to NYT’s analysis of the Equality of Opportunity database.


Universities that fall below the 17% of Pell recipients are usually selective, wealthy and private, according to the research report released by the Education Trust.
Among public universities, University of Missouri-Columbia(Mizzou) and Truman State University are placed in the bottom, with around 20% Pell students. Meanwhile, Mizzou and Washington University in St. Louis have around $12,000 net price for poor students, $2000 less than the state average level.


What made Mizzou and Washington University unfriendly to students from bottom quintile of income distribution? Interview

On the opposite, colleges that are less affluent and have less resources, such as College of the Ozarks, Baptist Bible College , and Ozark Christian College have more than 50% Pell recipients, and less than $10,000 net cost for students from low-income family.

College of the Ozarks, a Christian liberal-arts college in Missouri with around 400 million endowment, waived the tuition for student in exchange for working on campus 15 hours per week.
